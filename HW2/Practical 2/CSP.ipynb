```python
def is_empty(sudoku):
    """
    TODO: This function should check whether the sudoku is still empty (contains zero values)
    """
    count = 0
    for i in range(9):
        for j in range(9):
            if sudoku[i,j] == 0:
                return True
    return True
```

```python
def is_placement_valid(grid, num, position):
    """
    Determines whether it's valid to place the number 'num' at the given 'position' on the Sudoku grid.
    """
    row, col = position

    # Check the row
    for i in range(9):
        if grid[row][i] == num:
            return False

    # Check the column
    for i in range(9):
        if grid[i][col] == num:
            return False

    # Check the 3x3 subgrid
    start_row = (row // 3) * 3
    start_col = (col // 3) * 3
    for i in range(3):
        for j in range(3):
            if grid[start_row + i][start_col + j] == num:
                return False

    return True
```

```python
def backtrack(sudoku):
    """
    Solves the Sudoku puzzle using backtracking.
    """
    empty_position = None
    for i in range(9):
        for j in range(9):
            if(sudoku[i,j] == 0):
                empty_position = (i,j)
                break
    if not empty_position:
        return True
    row, col = empty_position
    for num in range(1, 10):
        if is_placement_valid(sudoku, num, empty_position):
            sudoku[row, col] = num 

            if backtrack(sudoku):
                return True
                     
            sudoku[row, col] = 0 
    
    return False
```

```python
def solve_sudoku(sudoku):
    """
    Solves the given Sudoku puzzle using OR-Tools CP-SAT Solver and returns the solved grid.
    
    :param sudoku: 2D numpy array representing the Sudoku puzzle (0 indicates empty cells)
    :return: Solved 2D numpy array or a message indicating failure
    """
    # Create the model
    model = cp_model.CpModel()
    cells = {}
    for i in range(9):
        for j in range(9):
            cells[(i, j)] = model.NewIntVar(1, 9, f'cell_{i}_{j}')

    for i in range(9):
        for j in range(9):
            if sudoku[i, j] != 0:
                model.Add(cells[(i, j)] == sudoku[i, j])

    for i in range(9):
        model.AddAllDifferent([cells[(i, j)] for j in range(9)])

    for j in range(9):
        model.AddAllDifferent([cells[(i, j)] for i in range(9)])

    for start_row in range(0, 9, 3):
        for start_col in range(0, 9, 3):
            subgrid = [
                cells[(start_row + i, start_col + j)]
                for i in range(3)
                for j in range(3)
            ]
            model.AddAllDifferent(subgrid)

    solver = cp_model.CpSolver()
    status = solver.Solve(model)

    if status == cp_model.FEASIBLE or status == cp_model.OPTIMAL:
        solved_grid = np.zeros((9, 9), dtype=int)
        for i in range(9):
            for j in range(9):
                solved_grid[i, j] = solver.Value(cells[(i, j)])
        return solved_grid
    else:
        return "No solution exists for the provided Sudoku puzzle."
```

### AC-3 Algorithm (20 Points)
First and foremost, implement the AC-3 algorithm to be later used in the problem

```python
def AC3(csp, neighbors):
    """
    csp: 
        dict{
            nodes: list of nodes,
            domains: dictionary {variable: [domain]},
            constraints: dictionary {(node_i, node_j): list of constraints
        },
    neighbors:
        dict {var: [neighbor]}
    
    TODO: Implement the AC3 algorithm
    
    return: a bool to show whether the problem is consistent and if so, the reduced domain of the problem
    """
    queue = []
    for x in neighbors.keys():
        if len(neighbors[x]) > 0:
            for y in neighbors[x]:
                queue.append((x, y))

    counts = []
    count_for = 0
    while len(queue) != 0:
        arc = queue.pop(0)
        res, csp_new, c = revise(csp, arc[0], arc[1])
        csp = csp_new
        if res:
            if len(csp["domains"][arc[0]]) == 0:
                counts.append(c)
                return False, csp["domains"]
            for Xk in neighbors[arc[0]]:
                count_for += 1
                if Xk != arc[1]:
                    queue.append((Xk, arc[0]))
        counts.append(max(c,count_for))
    print("AC3 require ", len(counts), " steps. Average steps:%.4g " %np.average(counts), ". Total operations: ", np.sum(counts))
    return True, csp["domains"]


def revise(csp, Xi, Xj):
    count = 0
    revised = False
    if (Xi, Xj) in csp["constraints"].keys():
        for Vi in csp["domains"][Xi]:
            satisfy = False
            for Vj in csp["domains"][Xj]:
                count += 1
                if (Vi, Vj) in csp["constraints"][(Xi, Xj)]:
                    satisfy = True
                    break
            if not satisfy:
                csp["domains"][Xi].remove(Vi)
                revised = True

    return revised, csp, count
```

### AC-4 Algorithm (20 Points)
Secondly, implement the AC-4 (Arc Consistency 4) algorithm. This algorithm enforces arc consistency by maintaining **support tables** and efficiently processing **a queue of arcs** for consistency checks. To help you implement it, here are the key steps and concepts to focus on:
1. **Support Tables**: For each pair of variable-value assignments, you must maintain a table that tracks the number of supporting values in the neighboring variables' domains. This table will allow you to quickly determine if a value is consistent with its neighbors
2. Start by initializing your support tables for all variable-value pairs. For each arc (X, Y), compute and store the number of values in Y's domain that support each value in X's domain based on the constraint between X and Y.
3. Initialize **a queue of arcs** with all pairs of variables that share a constraint
4. For each arc (X, Y) in the queue, check the support for each value in X's domain. If any value in X loses all supporting values in Y, remove that value from X’s domain.
5. If a value is removed from X’s domain, for every neighboring variable Z (other than Y), you must add the arc (Z, X) back into the queue for further consistency checks. This ensures that removing a value from X doesn’t create inconsistencies for other variables.
6. Whenever a value is removed from a domain, update the support tables for all neighboring variable-value pairs that depend on the removed value. This way, if a value of Z is only supported by the value that was removed from X, the support count for that value of Z should be decremented.
7. The algorithm terminates when the queue is empty, meaning all arcs have been processed and the domains of all variables are arc-consistent. If at any point a domain becomes empty, the problem is unsolvable.

```python
def AC4(csp):
    """
    AC-4 algorithm for arc consistency.

    :param csp: 
        dict {
            'nodes': list of nodes,
            'domains': dictionary {variable: [domain]},
            'constraints': dictionary {(node_i, node_j): function that returns True if satisfied}
        }
        
    :return: (bool, dict) - A boolean indicating if arc consistency was achieved, and the reduced domains
    """
    support = {}
    for x in csp["nodes"]:
        for val in csp["domains"][x]:
            support[(x, val)] = []

    deletion_stream = []
    counter = {}
    count = 0
    for (Xi, Xj) in csp["constraints"].keys():
        for Vi in csp["domains"][Xi]:
            tot = 0
            for Vj in csp["domains"][Xj]:
                count += 1
                if (Vi, Vj) in csp["constraints"][(Xi, Xj)]:
                    tot += 1
                    support[(Xj, Vj)].append((Xi, Vi))
            if tot == 0:
                csp["domains"][Xi].remove(Vi)
                deletion_stream.append((Xi, Vi))
                if len(csp["domains"][Xi]) == 0:
                    return False, csp["domains"]
            else:
                counter[(Xi, Xj, Vi)] = tot
    count2 = 0
    while len(deletion_stream) != 0:
        (xi, vi) = deletion_stream.pop(0)
        for (xj, vj) in support[(xi, vi)]:
            count2 += 1
            counter[(xj, xi, vj)] -= 1
            if counter[(xj, xi, vj)] == 0 and vj in csp["domains"][xj]:
                csp["domains"][xj].remove(vj)
                deletion_stream.append((xj, vj))
                if len(csp["domains"][xj]) == 0:
                    return False, csp["domains"]

    print("AC4 require ", max(count,count2), " steps")
    return True, csp["domains"]
```

### Conclusion
Explain the differences between these two algorithms in terms of time and space complexity. Which one do you prefer to use?

**Your answer**: 
1.  The time complexity of AC-3 is approximately $O(n^2d^3)$, This is because each arc may need to be rechecked multiple times if any domain changes. AC-3 uses less space than AC-4, with space complexity around $O(n^2)$ for storing the constraints and domains. 
2. The time complexity of AC-4 is better than AC-3 in cases with high inter-variable constraints. AC-4 achives $O(n^2d^2)$ by efficiently maintaining support tables that prevent redundant checks of variable assignments. This improvement makes AC-4 particularly useful when domains are large or there are many constraints between variables.


## Backtracking + AC-3 and AC-4 algorithm (20 Points)

```python
def is_consistent(var, value, assignment, csp, neighbors):
    for neighbor in neighbors[var]:
        if neighbor in assignment:
            if (value, assignment[neighbor]) not in csp["constraints"].get((var, neighbor), []):
                return False
    return True

def select_unassigned_variable(assignment, domains):
    unassigned_vars = [var for var in domains if var not in assignment]
    return min(unassigned_vars, key=lambda var: len(domains[var]))

def backtracking_search(csp, neighbors, use_ac3=True):
    assignment = {}
    original_domains = {var: list(domain) for var, domain in csp["domains"].items()}

    while True:
        if len(assignment) == len(csp["nodes"]):
            return assignment

        unassigned_vars = [v for v in csp["domains"] if v not in assignment]
        if not unassigned_vars:
            return None  

        var = min(unassigned_vars, key=lambda v: len(csp["domains"][v]))

        found_consistent_assignment = False

        for value in csp["domains"][var]:
            if all((value, assignment.get(neighbor)) in csp["constraints"].get((var, neighbor), [])
                   for neighbor in neighbors[var] if neighbor in assignment):

                assignment[var] = value

                local_csp = {
                    "nodes": csp["nodes"],
                    "domains": {v: list(d) for v, d in csp["domains"].items()},
                    "constraints": csp["constraints"]
                }
                local_csp["domains"][var] = [value]

                if use_ac3:
                    consistent, reduced_domains = AC3(local_csp, neighbors)
                else:
                    consistent, reduced_domains = AC4(local_csp)

                if consistent:
                    csp["domains"].update(reduced_domains)
                    found_consistent_assignment = True
                    break
                else:
                    csp["domains"] = {var: list(dom) for var, dom in original_domains.items()}
                    assignment.pop(var, None)

        if not found_consistent_assignment:
            if var in assignment:
                del assignment[var]
            else:
                return None  


csp = {
    "nodes": nodes,
    "domains": domains,
    "constraints": constraints
}

csp2 = copy.deepcopy(csp)

print("\nBacktracking + AC3 algorithm: \n")
start = time()
solution = backtracking_search(csp, graph, True)
s = time() - start

if solution:
    print("Solution found:", solution)
else:
    print("No solution found!")
    
print("Solution retrieved in: %.8g seconds" % s)

# TODO: Solve using AC4 algorithm

print("\nBacktracking + AC4 algorithm:\n")
start = time()
solution = backtracking_search(csp2, graph, False)
s = time() - start

if solution:
    print("Solution found:", solution)
else:
    print("No solution found!")
    
print("Solution retrieved in: %.8g seconds" % s)
```
